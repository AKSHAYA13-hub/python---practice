#### Introduction to Dictionaries

Dictionaries are unordered collection of items. They store data in key-value pairs. Keys must be unique and immutable(eg : strings, numbers, or tuples), while values can be of any type.



```python
## creating dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
students = {"name" : "Akshaya", "age" : 19, "grade" : "A"}
print(students)
print(type(students))
```

```python
## duplicate keys = do not mention key more than once
student = {"name" : "Akshaya", "age" : 19, "name" : "A"}
print(student)
```

```python
## accessing dictionary elements
student = {"name": "Akshaya", "age" : 19, "grade" : "A"}
print(student)
```

```python
##Accessing Dictionary Elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', 'Not found'))
```

```python
## Modifying Dictionary Elements
## dictionary are mutable, so you can add, update or delete elements

print(student)
```

```python
# Update value for the key 'age'
student["age"] = 20
print(student)

# added a new key and value pair
student["address"] = "123-Nian road"
print(student)
```

```python
## delete key and value pair
del student['grade']
print(student)
```

```python
## Dictionary Methods

keys = student.keys()    #get all keys in the dicitionary
print(keys)

values = student.values()  #get all values in the dictioanry
print(values)

items = student.items()  #get all key-value pairs as tuples
print(items)
```

```python
## Shallow copy = assigning one dictionary to another variable
student_copy = student
print(student)
print(student_copy)
```

```python
student['name'] = "Akshu"
print(student)
print(student_copy)
```

```python
## copy and shallow copy difference :
# shallow copy= shallow copy creates a new dictionary but references the same objects. 
# copy =  copy() method creates a new dictionary with copies of the objects in the original dictionary.
student_copy1 = student.copy()
print(student_copy1)
print(student)
```

```python
student['name'] = "achuu"
print(student)
print(student_copy1)
```

```python
##Iterating Over Dictionaries
## You can use loops to iterate over dictionaries, keys, values or items

#Iterating over keys
for key in student.keys():
    print(key)
```

```python
## Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
## Nested Dictionaries
students = {
    "student1": {
        "name" : "Akshaya",
        "age" :19,
        "city" : "hyderabad"
    },
    "student2" : {
        "name" : "Rahul",
        "age" : 14,
        "city" : "nizamabad"
    }
}
print(students)
```

```python
## Accessing Nested Dictionaries
print(students["student1"]["name"])
print(students["student2"]["age"])
```

```python
##Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")
```
