### DATA TYPES:


>> Data Type in python is a way to classify data items of a value that is holded by some variable.

>> They determine the type of opreations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data

 IMPORTANCE OF DATA TYPES IN PROGRAMMING:

>> Data types ensure that data is stored in an efficient way.

>> They help in performing corrct operations on data.

>> Proper use of data types can prevent errors and bugs in the program.

```python
## Integer example:

age = 34
type(age)
print(type(age))
```

```python
## Floating point example:


height = 5.9
print(height)
print(type(height))
```

### Arithemetic Operators:


```python
##  Arithementic Operations:

a = 10
b = 5

add_result = a + b  #addition
sub_result = a - b  #subtraction
mult_result = a * b  #multiplication
div_result = a / b  #division
floor_div_result = a // b  #floor division
moduluds_result = a % b  #modulus
exponent_result = a ** b  #exponentiation


print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(moduluds_result)
print(exponent_result)
```

#### Logical Operators


```python
## AND Operation:
# if both operands are true, the result is true
X = True
Y = True

result = X and Y
print(result)
```

```python
## OR operation:
# if at least one operand is true, the result is true
X = True
Y = False

result = X or Y
print(result)
```

```python
## NOT Operation:
X = True
not_X = not X
print(not_X)
```
