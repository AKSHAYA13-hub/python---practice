#####  List comprehension

Basics syntax              [expression for item in iterable]


with conditional logic     [expression for item in iterable if condition]


Nested List Comprehension  [expression for item1 in iterable1 for item2 in iterable2]

```python
### Basic list comprehension

square = [num**2 for num in range(10)]
print(square)
```

```python
## Lsit comprehension with condition

lst =[]
for i in range(10):
    if i%2 == 0:
        lst.append(i)
print(lst)
```

```python
even_numbers = [num for num in range(10) if num%2 == 0]
print(even_numbers)
```

```python
## Nested list comprehension

lst1 = [1, 2, 3, 4]
lst2 =['a', 'b', 'c', 'd']

pair = [[i, j] for i in lst1 for j in lst2]
print(pair)
```

```python
### List Comprehension with function calls

words  = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)
```

#### Coclusion
List comprehension area powerful and concise way to create lists in python. They are sytactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehension will help you write cleaner and more efficient python code.
