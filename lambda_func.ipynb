#### Lambda Functions in Python

Lambda function are small anonymous defined using the <b>lambda</b> keyword.They can have any no. of arguments but only one expression. Theyr are commonly used for short operations or as arguments to higher-order functions.

```python
## Syntax
lambda arguments: expression
```

```python
def addition(a, b):
    return a + b
addition(2, 3)
```

```python
## LAMBDA
addition = lambda a, b: a + b
print(type(addition))
print(addition(5, 5))
```

```python
def even(num):
    if num % 2 == 0:
        return True
    
even(24)
```

```python
even1 =lambda num: num%2==0
even(12)
```

```python
def add(x, y, z):
    return x+y+z

add(1, 2, 3)
```

```python
add1 = lambda x, y, z : x+y+z
add1(2, 2, 2)
```

```python
## map() - applies a function to all items in a list
numbers = [1, 2, 3, 4, 5, 6]
def square(number):
    return number**2
square(2)
```

```python
numbers = [1, 2, 3, 4, 5, 6]
list(map(lambda x: x**2, numbers))
```
