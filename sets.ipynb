#### SETS

Sets are built-in data type in pyhton used to store collection of unique elements. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries and performing mathematical set operations like union, intersection, difference and symmetric difference

```python
course = {"history", "math", 'physics', 'compsci'}
print(course)
```

```python
### create a set

my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_empty_set = set([1, 2, 3, 4, 5, 6])
print(type(my_empty_set))
print(my_empty_set)
```

```python
my_set = set([1, 2, 3,  2, 1, 4])
print(my_set)
```

```python
## Basics sets operation
##Adding and removing elements
my_set = set([1, 3, 4, 5, 6])
my_set.add(2)
print(my_set)
my_set.add(3)
print(my_set)
```

```python
## Removing elements from the set = It removes particualr element from the list
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

```python
my_set.discard(10)
print(my_set)
```

```python
## Pop method = It removes the first element from the set and return it
removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

```python
## Clear all elemenst from the set = clears the set
my_set.clear()
print(my_set)
```

```python
## set Membership test = Check if an element is present in the set
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operations 
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 5, 6, 7, 8}

# Union
union_set =  set1.union(set2)
print(f"Union: {union_set}")

# Intersection
intersection_set = set1.intersection(set2)
print(f"Intersection: {intersection_set}")


#Intersection_update = It updates the set1 with the intersection of set1 and set2
set1.intersection_update(set2)
print(set1)
```

```python
# Difference
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 5, 6, 7, 8, 9}
difference_set= set1.difference(set2)
print(f"Difference: {difference_set}")
```

```python
set2.difference(set1)
```

```python
## Symmetric Difference = It returns a set with elements from either set1 or set2 but not in both
set1.symmetric_difference(set2)
```

```python
## Sets Methods
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5}

## is subset = It checks whether set1 contains some or all elements of set2
print(set1.issubset(set2))

## is superset = It checks whether set1 contains all elements of set2
print(set1.issuperset(set2))
```

```python
lst = [1, 2, 2, 3, 4, 4, 5]
set(lst)
```

```python
## counting unique words in text

text = "In this tutorial we are discussing about sets"
words = text.split() #split = splits the text into list of words
print(words)
# convert list of words to set to get unique words

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
