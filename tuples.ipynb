#### Tuples

Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.


```python
## Creating a tuple

my_tuple = ()
print(my_tuple)
print(type(my_tuple))
```

```python
lst = list()
print(type(lst))
tp1 = tuple()
print(type(tp1))
```

```python
numbers = tuple([1, 3, 5, 7, 9])
numbers
```

```python
mixed_tpl = (1, "helllo", 3.14, True)
print(mixed_tpl)
```

```python
numbers = (1, 2, 3, 4, 5)
print(numbers[2])
print(numbers[-1])
print(numbers[::-1])
```

```python
## Tuple Operations

concatenation = numbers + mixed_tpl
print(concatenation)
```

```python
mixed_tpl*3

numbers*3
```

```python
## Immutabale Nature of Tuples
## Tuples are immutable, meaning their elements cannnot be changed once defined.

numbers[1] = 10
print(numbers)
```

```python
## Tuple methods

print(numbers.count(3))
print(numbers.index(3))
```

```python
## Packing and Unpacking

packed_tuple = 1, "hello", 3.14, False
print(packed_tuple)
```

```python
## Tuple unpacking

a, b, c, d = packed_tuple

print(a)
print(b)
```

```python
## Unpacking with asterisk (*) operator

values = (1, 2, 3, 4, 5, 6)
first, *middle, last = values
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples

lst = [ [1, 2, 3], [4, 5, 6], [7, 8, 9]]
lst[0][2]
```

```python
lst[2][0:3]
```

```python
nested_tpl = ((1, 2, 3), (4, 5, 6), (7, 8, 9))

## Accessing elements in nested tuples
print(nested_tpl[0])
print(nested_tpl[1][2])
```

```python
## Iterating over nested tuples
for sub_tuple in nested_tpl:
    for item in sub_tuple:
        print(item, end = " ")
    print()
```

#### Conclusion

Tuples are versatile and useful in many real-world scenarios wheere an immutable and ordered collection of items is required. They are commonly used in data structures, function arguments and return values and as dictionary keys. Understanding how to leverage tuples effectively can improve the efficiency and readability of your python code.
