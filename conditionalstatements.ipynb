#### Conditional Statements(if, elif, else)


```python
## if statement
# if statements are the conditional statements, it executes the block of code, if the condtion is true. otherwise, it don't

age = 20
if age >= 18:
    print("You are allowed to vote in the elections")
```

```python
age >= 12
```

```python
## else
## The else statement executes a block of code if the condition in the if statements is False.

age = 16
if age >= 18:
    print("You are eligible to vote")

else:
    print("you are a minor")
```

```python
## elif
## The elif statement allows you to check multiple-conditions, It stands for "else - if"
age = 17

if age < 13:
    print("You are a child")
elif age < 18:
    print("You are a teenager")
else:
    print("you are an adult")
```

```python
## Nested Conditional Statements

## You can place one or more if, elif, or else statements inside another if, elif, else statements

## number even, odd, negative

num = int(input("Enter the number"))

if num > 0:
    print("The number is positive")
    if num%2 == 0 :
        print("The number is even")
    else:
        print("The number is odd")
    
else:
    print("The number is zero or negative")
```

```python
## Practical Examples
## Determine if a year is a leap year using nested condition statements

year = int(input("Enter the year"))

if year%4 == 0 :
    if year%100 == 0:
        if year%400 == 0:
            print(year, "is a leap year")
        else:
            print(year, "is not a leap year")
    else:
        print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```
