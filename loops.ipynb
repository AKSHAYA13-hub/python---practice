```python
### Check Armstrong Number

num = int(input("Enter a number: "))
temp = num
sum = 0

digits = len(str(num))

while temp>0:
    digit = temp % 10
    sum = sum + (digit ** digits)
    temp = temp//10

if sum == num:
    print(f"{num} is an armstrong number")

else:
   print(f"{num} is not an armstrong number") 
```

#### For Loop:

```python
nums = [1, 2, 3, 4, 5]
for num in nums:
    
    print(num)
```

```python
## Break statement

num = [1, 2, 3, 4, 5, 6]

for num in nums:
    if num == 3:
        print("Found")
        break
    print(num)
```

```python
## Continue statement


for num in nums:
    if num == 3:
        print("Found")
        continue
    print(num)
```
