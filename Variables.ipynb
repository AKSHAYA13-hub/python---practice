Variables

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
### Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
diff = num1 - num2
product = num1 * num2
quoitent = num1 / num2

print("Sum: ", sum)
print("Difference: ", diff)
print("product: ", product)
print("Quotient: ", quoitent)
```

```python
## Count Method:
## It is used to count the no. of words does the variable contains
message = "Hello World"
print(message.count(message))
print(message.count("Hello"))
print(message.count("World!"))
print(message.count("l"))
print(message.count("o"))
```

```python
## Find
# Use to find the 
message = "Hello World"
print(message.find('World'))
print(message.find('Hello'))
print(message.find('l'))
print(message.find('Universe'))
```

```python
## Replace:
# It is used to replace a string or character or value, from the given one
message = "Hello World"
new_msg = message.replace('World', 'Universe')
print(message)
print(new_msg)
```

```python
## Concatenate
greeting = "Hello"
name = "Akshaya"
message = greeting + name
print(message)

# to add space in between 2 strings is:
message1 = greeting + ',' +' ' +name
print(message1)

message1 = greeting + ',' +' ' +name +'. Welcome'
print(message1)

message2 = '{}, {}. Wlc!'.format(greeting, name)
print(message2)
```

```python
## formatted strings:

greeting = "Hello"
name = "Akshaya"
message = f'{greeting}, {name.upper()}.wlc!'
print(message)
```
